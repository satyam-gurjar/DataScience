```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"/content/anime.csv")
```

```python
def extract_episodes(txt):
  check = False
  data = ""
  for i in txt:

    if i == ')':
      break

    if i == '(':
      check = True
      continue

    if check == True:
      data += i
  return data
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
df['Episodes']=df['Episodes'].str.replace(" eps","")
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```

```python
df
```

```python
df.loc[0]['Title']
```

```python
def extraction_time(txt):
  data = ""
  for i in range(len(txt)):
    if txt[i] == ')':
      for j in range(i+1,i+20):
        data += txt[j]
      return data
```

```python
df['Total Time']=df['Title'].apply(extraction_time)
```

```python
df
```

```python
df[['Start', 'End']] = df['Total Time'].str.split(' - ', expand=True)

# convert to datetime
df['Start'] = pd.to_datetime(df['Start'], format='%b %Y')
df['End'] = pd.to_datetime(df['End'], format='%b %Y')

# calculate total months
df['Months'] = (df['End'].dt.year - df['Start'].dt.year) * 12 + \
                     (df['End'].dt.month - df['Start'].dt.month)
```

```python
df
```

```python
# remove everything after TV/Movie/OVA/ONA
df['Title'] = df['Title'].str.replace(
    r'(TV|Movie|OVA|ONA).*', '', regex=True
)

# strip extra spaces
df['Title'] = df['Title'].str.strip()
```
