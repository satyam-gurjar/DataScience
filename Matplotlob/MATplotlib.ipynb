```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib **Basics**

```python
x = np.linspace(0,5,11)
x
y = x **2
y
```

```python
plt.plot(x,y)
plt.title('Hello')
plt.xlabel("X")
plt.ylabel("Y")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(y**2,x**3)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,y**2)
```

**Object Oriented Way**

```python
#fig = plt.figure()
fig = plt.figure(figsize=(12,8),dpi=100)
axis1 = fig.add_axes([0.1,0.1,1,1])

axis1.plot(x,y)

axis2 = fig.add_axes([0.3,0.6,0.3,0.3])
axis2.plot(x,y)
```

**Types of Plot**

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

**Saving Plots**

```python
fit = plt.figure()
ax=fit.add_axes([0.1,0.1,0.6,0.7])
ax.plot(x,y)
plt.savefig("basic.png")
```
